```python
import sympy 
import matplotlib.pyplot as plt
import numpy as np
from methodLUP import solve_LUP
from polynomial import Polynomial
%matplotlib inline
```

```python
x_vec = [0.1, 0.5, 0.9, 1.3, 1.7, 2.1]
y_vec = [-2.3026, -0.69315, -0.10536, 0.26236, 0.53063, 0.74194]
```

```python
k = 1
pol = Polynomial(k)
pol.make(y_vec,x_vec)
print(f'Y = {y_vec}')
print(f'X = {x_vec}')
print(f'Полином {k}-й степени:')
pol.print_polynom()
print(f'Сумма квадратов ошибок = {pol.mse}')
x_pol = np.arange(x_vec[0]-1,x_vec[-1]+1,0.1)
y_pol = list(map(lambda xi: pol.calc_point(xi),x_pol))

plt.title('Графики полинома и табличных значений функции')

plt.scatter(x_vec,y_vec, color = "blue", label='y')
plt.plot(x_pol,y_pol,color = "red", label=f'F{k}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
